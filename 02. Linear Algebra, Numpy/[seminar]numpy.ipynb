The second way to create is using the built-in functions __`numpy.eye (N, M = None, ...) `__, __`numpy.zeros (shape, ...)` __, __`numpy.ones (shape, ...) `__.

The first function creates a unit matrix of size $ N \ times M $; if $ M $ is not specified, then $ M = N $.

The second and third functions create matrices consisting entirely of zeros or ones, respectively. As the first argument, you must specify the dimension of the array - a tuple of integers. In the two-dimensional case, this is a set of two numbers: the number of rows and columns of the matrix.

__Examples:__

```python
b = np.eye(5)
print("Unit matrix:\n", b)
```

```python
c = np.ones((7, 5))
print("Matrix consisting of one units:\n", c)
```


__Please note: the dimension of the array is not given by two function arguments, but by one - a tuple!__

So - __`np.ones (7, 5) `__ - create an array does not work, because the function as a parameter` shape` is passed to `7`, and not the tuple` (7, 5) `.


And finally, the third method is using the __`numpy.arange function ([start,] stop, [step,], ...) `__, which creates a one-dimensional array of consecutive numbers from the interval __` [start, stop)` __ with a given step __`step`__, and _methoda_ __`array.reshape (shape) `__.

The __`shape`__ parameter, as in the previous example, specifies the dimension of the matrix (a tuple of numbers). The logic of the method is clear from the following example:


__The memory of the theory .__

__Transported matrix__ $ A ^ {T} $ is the matrix obtained from the original $ A $ matrix by replacing rows with columns. Formally: the elements of the matrix $ A ^ {T} $ are defined as $ a ^ {T} _ {ij} = a_ {ji} $, where $ a ^ {T} _ {ij} $ is an element of the matrix $ A ^ {T } $, standing at the intersection of row number $ i $ and column number $ j $.

```python
a = np.array([[1, 2], [3, 4]])
b = np.transpose(a)
c = a.T.T
```

```python
print("Matrix:\n", a)
print("Transpose function:\n", b)
print("Transposing method:\n",  c)
```


__The memory of the theory .__ The __ multiplication operation __ is defined for two matrices, such that the number of columns of the first is equal to the number of rows of the second.

Let the matrices $ A $ and $ B $ be such that $ A \ in \ mathbb {R} ^ {n \ times k} $ and $ B \ in \ mathbb {R} ^ {k \ times m} $. __The product of __ matrices of $ A $ and $ B $ is the matrix $ C $ such that $ c_ {ij} = \ sum_ {r = 1} ^ {k} a_ {ir} b_ {rj} $, where $ c_ {ij} $ Is an element of the matrix $ C $ standing at the intersection of the row with the number $ i $ and the column with the number $ j $.

In `NumPy`, the matrix product is calculated using the __`numpy.dot function (a, b, ...)` __ or using the _ method_ __`array1.dot (array2) `__, where __`array1`__ and __` array2`__ - multiply matrices.

* Optimizing (extremums):

```python
from scipy.optimize import minimize
```

```python
?minimize
```

Minimizing fuction $x^2$

```python
def f(x):
    return x ** 2
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = np.arange(-3, 3, .1)
y = f(x)

plt.plot(x,y)
plt.show()
```

```python
res = minimize(f, x0=100)
```

```python
res
```

Here you need to look at 4 lines: `fun, message, success and x`
`fun` - functions at minimum
`message` - a service message about the end of the process (maybe" successful ", as here, or a message that something went wrong (" did not converge "))
`success` - True, if it is successfully met (but it is always better to still watch and` message`)
`x` - the point at which the minimum is reached


You can give different keywords to the `method` argument to change the optimization method, it’s better to read the documentation or google it in more detail.

* Integration:

```python
from scipy.integrate import quad, odeint
from scipy.special import erf
```

```python
def f(x):
    return np.exp(-x ** 2)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = np.arange(-3, 3, .1)
y = f(x)

plt.plot(x,y)
plt.show()
```


Adaptive numerical integration (may be to infinity). `err` - error evaluation.

```python
res, err = quad(f, 0, np.inf)
print(np.sqrt(np.pi) / 2, res, err)
```

```python
res, err = quad(f, 0, 1)
print(np.sqrt(np.pi) / 2 * erf(1), res, err)
```
